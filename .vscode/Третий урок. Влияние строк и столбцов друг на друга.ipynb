Групировка, сортировка и корреляция

```python
import pandas as pd
df= pd.read_csv(r'C:\Users\nikry\fintech_analysis\students.csv', delimiter = ',') 
df.info()
```

```python
df_cut = df[['Age', 'Growth', 'Weight']]
df_cut.head()
```

Сортировка

```python
df_cut.sort_values(by = 'Age', ascending=False)
#Отсортировали таблицу по убыванию возраста 
```

```python
df_cut.sort_values('Growth', ascending = True)
#Отсортировали таблицу по увеличению роста 
```

```python
df_cut.sort_values(['Age', 'Growth'], ascending = [False, False])
#Отсортировали таблицу по убыванию возраста и убыванию роста(в случае если у разных студентов возраст будет одинаковым)
```

```python
df_cut.sort_values(['Growth', 'Age'], ascending = [False, False])
#Отсортировали таблицу по убыванию роста и убыванию возраста(в случае если  у разных студентов рост будет одинаковым)
```

```python
df_sorted = df_cut.sort_values('Growth', ascending = False)
```

```python
df_sorted.iloc[3,] 
#Обратились по уникальному индексу (номеру строки), то есть вывели данные о четвёртом по счёту студенте 
```

```python
df_sorted.iloc[3,0] 
#Вторым параметром задали вывод определённого столбца (в данном случае столбца с признаком возраста)
```

Добавление нового стоблца

```python
df_cut['Mass index'] = 10000*df_cut['Weight'] / (df_cut['Growth'] * df_cut['Growth'])
#Добавили столбец с индексом массы тела
```

```python
df_cut.head()
```

```python
df_cut.sort_values('Mass index', ascending = False)
```

Коэффициент корреляции (КК)

```python
df_cut.corr()
```

Решение практических задач по третьему уроку 

```python
df.info()
```

```python
df_cut2 = df[['Year of birth', 'Growth', 'Weight', 'Hair length', 'Shoe size']]
```

```python
df_cut2.sort_values('Shoe size', ascending = True)
```

```python
df_cut2.sort_values('Shoe size', ascending = False)
```

```python
df_cut2.sort_values(['Shoe size', 'Growth'], ascending = [True, True]).head(9)
```

```python
df_cut2.sort_values(['Shoe size', 'Growth'], ascending = [True, False])
```

```python
df_cut2.sort_values(['Shoe size', 'Growth'], ascending = [False, True])
```

```python
df_cut2['Full length'] = df_cut2['Growth'] + df_cut2['Hair length']
max_length = df_cut2['Full length'].max()
print(f'Максимальное значение признака равно {round(max_length)} см')
```

```python
df_cut2.corr()
```
