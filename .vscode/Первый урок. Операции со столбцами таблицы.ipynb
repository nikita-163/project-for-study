Задача. У вкладчика было 15600 рублей. Он положил их в банк под 9% годовых. Спустя год он переложил все свои деньги в другой банк под 16% годовых. Сколько всего денег у него будет к концу второго года. 

```python
#ответ
15600 * 0.09 * 0.16
```
