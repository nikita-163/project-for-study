```python
df1 = df[df['Growth'] < 165]
df2 = df[(df['Growth'] >= 165) & (df['Growth'] <= 175)]
df3 = df[df['Growth'] > 175]
#разделили студентов по росту на 3 группы:
#1) студенты с ростом больше 165
#2) студенты с ростом 165 и больше и студеннты с ростом 175 и меньше
#3) студенты с ростом больше 175
```

```python
df1['Growth'].mean(), df2['Growth'].mean(), df3['Growth'].mean()
```
