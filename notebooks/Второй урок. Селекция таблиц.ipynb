```python
dor_lovers = df[df['Animal'] == 'Собак']
floor1 = dor_lovers[dor_lovers['Floor number'] == 1]
percent = round(len(floor1) / len(dor_lovers) * 100)
print(f'Способ 1: процент живущих на 1 этаже = {percent}%')
```

```python
all_people = dor_lovers.shape[0]
floor1_2 = dor_lovers[dor_lovers['Floor number'] == 1].shape[0]
percent2 = round(floor1_2 / all_people * 100)
print(f'Способ 2: процент живущих на 1 этаже = {percent2}%')
```

```python
dor_lovers = df[df['Animal'] == 'Собак']
percent3 = round((dor_lovers['Floor number'] == 1).mean() * 100)
print(f'Способ 3: процент живущих на 1 этаже = {percent3}%')
```

```python
percent = (df['Maths rating'] > 60).mean() * 100
print(f'{round(percent)}% студентов с рейтингом по математике выше 60 баллов')
```

```python
df['Brother-sister'].value_counts()
```

```python
single_child = df[df['Brother-sister'] == 'нет ни брата ни сестры']
median1 = single_child['Social network duration min'].median()
print(f'Длительность {round(median1)}')
```

```python
not_single_child = df[(df['Brother-sister'] == 'есть брат') | (df['Brother-sister'] == 'есть сестра')]
median2 = not_single_child['Social network duration min'].median()
print(f'Длительность {round(median2)}')
```

```python
difference_duration = round(median1 - median2)
print(f'Разница составляет {difference_duration} минут')
```

```python
KFC_persons = df[df['Fastfood'] == 'KFC']
avg_KFC_weight = KFC_persons['Weight'].mean()
print(f'Средний вес {round(avg_KFC_weight)} кг')
```

```python
Burger_persons = df[df['Fastfood'] == 'Бургер кинг']
avg_Burger_weight = Burger_persons['Weight'].mean()
print(f'Средний вес {round(avg_Burger_weight)} кг')
```

```python
Mac_persons = df[df['Fastfood'] == 'Макдональдс (или как он там сейчас называется?)']
avg_Mac_weight = Mac_persons['Weight'].mean()
print(f'Средний вес {round(avg_Mac_weight)} кг')
```

```python
dif_weight = round(avg_KFC_weight - avg_Mac_weight)
print(f'Средний вес студентов, употребляющих KFC, больше на {dif_weight} кг')
```
