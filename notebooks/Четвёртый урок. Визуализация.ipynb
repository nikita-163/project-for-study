```python
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv(r'C:\Users\nikry\project-for-study\data\students.csv', delimiter = ',')
df.info()
```

Построение визуализаций 

```python
sns.displot(data = df, x = 'Growth')
#Построили гистограмму на основе критерия роста
#Можно сделать вывод, что наибольшее количество студентов ростом 165 см
```

```python
sns.displot(data = df, x = 'Growth', kind = 'kde') 
```

```python
sns.scatterplot(data = df, x = 'Growth', y = 'Weight')
#Построили диаграмму рассеяния на основе критерия роста и веса
#Можно сделать вывод, что чем больше у студента рост, тем больше его вес
```

```python
sns.scatterplot(data = df, x = 'Growth', y = 'Weight', hue = 'Sex')
#Построили диаграмму рассеяния на основе критерия роста, веса и задали категориальный параметр
#Дополняя предыдущий вывод можно сказать, что мужчины обладают большим весом и ростом чем женщины
```

```python
sns.scatterplot(data = df, x = 'Growth', y = 'Weight', hue = 'Coin')
#Точки на диаграмме рассеивания расположены хаотично, так как нет прямой зависимости между биологическими признаками и тем, какая сторона монетки выпадает 
```

```python
sns.countplot(data = df, x ='Sex')
#Построили счётную диаграмму на основе критерия пола
#Можно сделать вывод, что среди людей, прошедших опрос, женщин почти в два раза больше
```

```python
sns.countplot(data = df, x ='Sex', hue = 'Animal')
#Построили счётную диаграмму на основе критерия пола и задали категориальный параметр, который делит столбцы внутри своей группы
#Дополняя предыдущий вывод можно сказать, что в обеих группах любимым животным является кошка 
```

```python
sns.countplot(data = df, x ='Animal', hue = 'Sex')
#Поменяли значения праметров друг с другом и выяснили, что кошек и собак больше любят женщины, а ненавидят этих животных примерно равное количество голосовавших
```

```python
df_cut = df[['Age', 'Growth', 'Weight', 'Shoe size', 'Sex']]
df_cut.head()
```

```python
sns.pairplot(df_cut)
#Построили парную диаграмму на основе критериев пола(без указания доп параметра не учитывается), возраста, роста, веса и размера обуви
```

```python
sns.pairplot(df_cut, hue = 'Sex')
#Построили парную диаграмму на основе критериев пола, возраста, роста, веса и размера обуви и задали категориальный параметр пола
```
