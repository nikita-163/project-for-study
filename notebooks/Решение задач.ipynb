```python
import pandas as pd
df = pd.read_csv(r'C:\Users\nikry\project-for-study\data\students.csv', delimiter = ',')
df.info()
```

Задача 1. 
Найти студентов, которые одновременно соответствуют трём условиям

```python
df1 = df[(df['Hostel'] == 'да') & (df['Sex'] == 'женский') & (df['Russian rating'] > 85)] 
avg_weight = df1['Weight'].mean()
print(f'Количество строк {df1.shape[0]}, средний вес {round(avg_weight)} кг')
```

Задача 2. Любители каких животных лучше всего знают математику?

```python
df2 = df.groupby('Animal')['Maths rating'].mean().round(1).sort_values(ascending=False)
df2.head()
```

```python
df2 = df.groupby('Animal')[['Maths rating', 'Russian rating']].mean().round(1).sort_values(['Maths rating'], ascending=False)
df2.head()
```
